```python
import numpy as np
import matplotlib.pyplot as plt

# Fixed parameters
p = 0.45         # Probability of winning each bet
B = 10           # Initial budget
n_simulations = 1000

# List of target capitals
target_capitals = [20, 30, 40, 50, 60, 70]

win_rates = []

for C in target_capitals:
    wins = 0
    
    for _ in range(n_simulations):
        capital = B
        # Keep betting until ruin (0) or reaching target C
        while capital > 0 and capital < C:
            if np.random.rand() < p:
                capital += 1
            else:
                capital -= 1
        
        # Check if we ended at the target
        if capital == C:
            wins += 1
    
    # Record the fraction of times we reached the target
    win_rate = wins / n_simulations
    win_rates.append(win_rate)

# Plot the relationship between target capital and win rate
plt.figure(figsize=(8, 5))
plt.plot(target_capitals, win_rates, marker='o')
plt.title("Gambler's Ruin: Win Rate vs. Target Capital (p = 0.45, B = 10)")
plt.xlabel("Target Capital (C)")
plt.ylabel("Win Rate")
plt.grid(True)
plt.show()
```
